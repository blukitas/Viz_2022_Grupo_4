```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sqlalchemy import create_engine
import os
```

```python
file_path = "/home/lucas/Downloads/Viz/Viz_2022_Grupo_4/docker/challenge_data/"
file_folder = "Activity Logs/"
file_folder_postproc = "fix_activity_logs/"
```

```python
for file in os.listdir(file_path + file_folder):
    split_file_name = file.split(".")
    # print(split_file_name[-1:])

    if split_file_name[-1:] == ["csv"]:
        print(file)
        df = pd.read_csv(file_path + file_folder + "/" + file)
        df["timestamp"] = pd.to_datetime(df["timestamp"])

        df_preproc = (
            df.sort_values(by=["timestamp"], ascending=True)
            .groupby(by=[x for x in df.columns if x != "timestamp"])
            .agg({"timestamp": [min, max]})
            .reset_index()
        )

        # Create columns, and remove the multiindex column
        df_preproc["ts_min"] = df_preproc["timestamp"]["min"]
        df_preproc["ts_max"] = df_preproc["timestamp"]["max"]
        del df_preproc["timestamp"]

        # In this point columns is a tuple. Reduce to the first level
        df_preproc.columns = df_preproc.columns.get_level_values(0)

        # Save, without index, to another folder
        df_preproc.to_csv(
            f"{file_path + file_folder_postproc}/{split_file_name[0]}.csv", index=False
        )
```
