```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sqlalchemy import create_engine
```

```python
POSTGRES_USER = "viz_user"
POSTGRES_PASSWORD = "viz_pass"
POSTGRES_DB = "viz"
```

```python
engine = create_engine(
    f"postgresql+psycopg2://{POSTGRES_USER}:{POSTGRES_PASSWORD}@localhost/{POSTGRES_DB}"
)
```

```python
SQL = """select 
	"month", 
	educationlevel, -- Podemos cambiar el nivel de agrupamiento
	ROUND(avg(wage_amt)::numeric, 2) as avg_wage_amt, 
	ROUND(avg(food_amt)::numeric, 2) as avg_food_amt, 
	ROUND(avg(education_amt)::numeric, 2) as avg_education_amt, 
	ROUND(avg(shelter_amt)::numeric, 2) as avg_shelter_amt
from 
	vo_monthly_wage_cohorts vmwc 
group by 1,2
order by 1,4
"""
```

```python
df = pd.read_sql(SQL, engine)
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
# figure(figsize=(20, 10), dpi=80)

# multiple line plots
ax.plot(
    "month",
    "avg_food_amt",
    data=df[df.educationlevel == "Graduate"],
    marker=".",
    markerfacecolor="blue",
    markersize=12,
    color="skyblue",
    linewidth=4,
)
ax.plot(
    "month",
    "avg_food_amt",
    data=df[df.educationlevel == "Bachelors"],
    marker="o",
    markerfacecolor="red",
    markersize=12,
    color="red",
    linewidth=4,
)
ax.plot(
    "month",
    "avg_food_amt",
    data=df[df.educationlevel == "HighSchoolOrCollege"],
    marker="^",
    markerfacecolor="green",
    markersize=12,
    color="green",
    linewidth=4,
)
ax.plot(
    "month",
    "avg_food_amt",
    data=df[df.educationlevel == "Low"],
    marker="v",
    markerfacecolor="black",
    markersize=12,
    color="black",
    linewidth=4,
)
# plt.plot( 'month', 'avg_food_amt', data=df, marker='', color='olive', linewidth=2)
# plt.plot( 'month', 'avg_food_amt', data=df, marker='', color='olive', linewidth=2, linestyle='dashed', label="toto")
ax.legend(["Graduate", "Bachelors", "HighSchool", "Low"])
plt.show()
```

# financial journal


```python
SQL = """select * from vo_monthly_wage"""
df = pd.read_sql(SQL, engine)
```

```python
df.spend_amt = -1 * df.spend_amt
```

```python
fig, ax = plt.subplots(figsize=(12, 5))

for participant in df.participantid.tolist():
    ax.plot(
        "month",
        "earning_amt",
        data=df[df.participantid == participant],
        marker=".",
        markerfacecolor="green",
        markersize=1,
        color="lightgreen",
        alpha=0.1,
        linewidth=1,
    )
    ax.plot(
        "month",
        "spend_amt",
        data=df[df.participantid == participant],
        marker=".",
        markerfacecolor="red",
        markersize=1,
        color="red",
        alpha=0.1,
        linewidth=1,
    )
# ax.legend(["Graduate", "Bachelors", "HighSchool", "Low"])
plt.show()
```
